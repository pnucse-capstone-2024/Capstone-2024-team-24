```python

def make_csv(target_folder,csvfilename):
    df = pd.DataFrame()
    for (root,directories,files) in os.walk(target_folder):
        for file in files:

            with open(os.path.join(target_folder,file),'r',encoding='UTF-8') as f:
                data = json.load(f)
                user = data['user']
                item = data['item']
                item_ = {'image': item['imgName'],'era': item['era'],'style': item['style'],'item_gender': item['gender'],**item['survey']}

                user.update(item_)
                new_df = pd.DataFrame([user])
                df = pd.concat([df, new_df], ignore_index=True)
    
    df_copy=df.copy()
    df['r_gender']=df['r_gender'].replace({1:'남성',2:'여성'})
    df['age']=df['age'].replace({1:'20대',2:'30대',3:'40대',4:'50대'})
    df['mar']=df['mar'].replace({1:'미혼',2:'기혼'})
    df['job']=df['job'].replace({1:'전업주부',2:'기술/전문직',3:'판매/서비스직',4:'사무/관리직',5:'학생',6:'기타'})
    df['income']=df['income'].replace({1:'200만원 미만',2:'200만원대',3:'300만원대',4:'400만원대',5:'500만원대',6:'600만원이상'})
    df['r_style1']=df['r_style1'].replace({1:'화려하고 독특한',2:'무난하고 평범한'})
    df['r_style2']=df['r_style2'].replace({1:'남성적/여성적인',2:'중성적인'})
    df['r_style3']=df['r_style3'].replace({1:'전통적인',2:'트랜디한'})
    df['r_style4']=df['r_style4'].replace({1:'포멀한',2:'캐주얼한'})
    df['r_style5']=df['r_style5'].replace({1:'활발한',2:'점잖은'})
    df['item_gender']=df['item_gender'].replace({'M':'남성','W':'여성'})
    df['Q2']=df['Q2'].replace({1:'봄/가을',2:'여름',3:'겨울'})
    df['Q3']=df['Q3'].replace({1:'출근',2:'데이트',3:'공식행사',4:'사교모임',5:'일상생활',6:'레저스포츠',7:'여행/휴가',8:'기타'})
    df['Q411']=df['Q411'].replace({1:'헐렁함',2:'적당함',3:'타이트함'})
    df['Q412']=df['Q412'].replace({1:'어두움',2:'밝음'})
    df['Q413']=df['Q413'].replace({1:'차가움',2:'따듯함'})
    df['Q414']=df['Q414'].replace({1:'무거움',2:'가벼움'})
    df['Q5']=df['Q5'].replace({1:0,2:1})
    
    df['Q5'] = df['Q5'].astype(bool)

    for i in range(1,16+1):
        num = format(i,'02')
        col = 'Q42' + num
        df[col] = df[col].astype(bool)


    df=df.rename(columns={'Q1':'선호여부','Q2':'어울리는계절','Q3':'적합한착용상황','Q411':'핏평가','Q412':'색깔','Q413':'색감','Q414':'분위기','Q4201':'멋있다','Q4202':'도시적이다','Q4203':'트렌디하다','Q4204':'세련되다','Q4205':'깔끔하다','Q4206':'화려하다','Q4207':'독특하다','Q4208':'무난하다','Q4209':'개방적이다','Q4210':'실용적이다','Q4211':'활동적이다','Q4212':'편안하다','Q4213':'발랄하다','Q4214':'여성적이다','Q4215':'남성적이다','Q4216':'부드럽다','Q5':'스타일선호'})
    df_copy=df_copy.rename(columns={'Q1':'선호여부','Q2':'어울리는계절','Q3':'적합한착용상황','Q411':'핏평가','Q412':'색깔','Q413':'색감','Q414':'분위기','Q4201':'멋있다','Q4202':'도시적이다','Q4203':'트렌디하다','Q4204':'세련되다','Q4205':'깔끔하다','Q4206':'화려하다','Q4207':'독특하다','Q4208':'무난하다','Q4209':'개방적이다','Q4210':'실용적이다','Q4211':'활동적이다','Q4212':'편안하다','Q4213':'발랄하다','Q4214':'여성적이다','Q4215':'남성적이다','Q4216':'부드럽다','Q5':'스타일선호'})

    
    df.to_csv(csvfilename+'.csv',index=False)
    df_copy.to_csv(csvfilename+'_encode.csv',index=False)

    return df,df_copy
```

```python
df,df_copy=make_csv('./Training/TL_man_2019','TL_man_2019')

df.head()
```
